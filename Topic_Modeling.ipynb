```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
articles = pd.read_csv("/content/npr.csv")
articles.info
```

```python
cv = CountVectorizer(max_df =0.95, min_df = 2, stop_words='english')
dtm = cv.fit_transform(articles['Article'])
dtm.shape
#dtm is document term matrix
```

```python
LDA = LatentDirichletAllocation(n_components=7, random_state=42)
topic_results = LDA.fit_transform(dtm)
LDA.components_.shape
```

```python
for i,arr in enumerate(LDA.components_):
  print(f'Top 15 words for topic#{i}')
  print([cv.get_feature_names_out()[i]for i in arr.argsort()[-15:]])
  print('\n\n')
```

```python
articles['topic'] = topic_results.argmax(axis=1)
articles
```
